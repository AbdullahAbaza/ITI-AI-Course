```python
## import packages
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
```

```python
driver = webdriver.Edge()
```

```python
# make a request
driver.get("https://www.amazon.eg/s?rh=n%3A21832883031&fs=true&language=en&ref=lp_21832883031_sar")
```

```python
products = driver.find_elements("xpath", '//span[@class = "a-size-base-plus a-color-base a-text-normal"]')
prices = driver.find_elements("xpath", '//span[@class="a-price-whole"]')

for product in products:
    print(product.text)

for price in prices:
    print(price.text)
```

```python
import urllib
df=pd.DataFrame([],columns=['Product','Price','Image_Url','Total_Ratings','Average_Rating(5)'])

```

```python
driver = webdriver.Edge()
# implicit wait for 5 seconds
driver.implicitly_wait(5)
# maximize with maximize_window()
driver.maximize_window()
# make a request
driver.get("https://www.amazon.eg/s?rh=n%3A21832883031&fs=true&language=en&ref=lp_21832883031_sar")
```

```python
for i in range(5):
    products_Sections = driver.find_elements(By.XPATH,"//div[@class='a-section a-spacing-base']")

    for section in products_Sections:
        try:
            title=section.find_element(By.XPATH,".//span[@class='a-size-base-plus a-color-base a-text-normal']")
            print(title.text)

            price=section.find_element(By.XPATH,".//span[@class='a-price-whole']")
            print(price.text)

            image_src=section.find_element(By.XPATH,".//img[@class='s-image']")
            url=image_src.get_attribute('src')
            print(url)

            rating=section.find_element(By.XPATH,".//div[@class='a-row a-size-small']/span").get_attribute('aria-label')
            print(rating.split()[0])

            total_rating=section.find_element(By.XPATH,".//span[@class='a-size-base s-underline-text']")
            print(total_rating.text)
        
            
            if (np.double(price.text.replace(",","")) > 1300):
                # urllib.request.urlretrieve(url,"{}.jpg".format(title.text))
                df.loc[len(df)] =[title.text,price.text,url,total_rating.text,rating.split()[0]]
        except:
            print('error')

    # identify element with link text and click()
    link=driver.find_element(By.LINK_TEXT, "Next")
    link.click()
```
